```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

effect_quest = pd.read_csv("C:/Users/theas/OneDrive - Région Île-de-France/Bachelor 2/Code python/Pandas python/effec1.quest.compil.csv", encoding="latin-1")
usage_effect = pd.read_csv("C:/Users/theas/OneDrive - Région Île-de-France/Bachelor 2/Code python/Pandas python/usages.effec1.csv", encoding="latin-1")
```

```python
merged_df = pd.merge(effect_quest, usage_effect, on='Student_ID')
```

```python
merged_df.head()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)


print(merged_df.columns)
```

```python
columns_to_sum = ["S1.L1", "S1.L2", "S1.L3", "S1.L4", "S1.L5", "S2.L1", "S2.L2", "S2.L3", "S2.L4", "S2.L5", "S3.L1.1", "S3.L1.2", "S3.L2", "S3.L3", "S3.L4", "S3.L5", "S4.L1.1", "S4.L1.2", "S4.L2", "S4.L3", "S4.L4", "S4.L5", "S5.L1.1", "S5.L1.2", "S5.L2", "S5.L3", "S5.L4", "S5.L5"]
columns_to_keep = ['Student_ID'] + ["Gender"] + ["Country_HDI"] + columns_to_sum + ["Quizz.1.bin", "Quizz.2.bin", "Quizz.3.bin", "Quizz.4.bin", "Quizz.5.bin"]
filtered_df = merged_df[columns_to_keep]
```

```python
filtered_df.fillna(0)
filtered_df.head()
```

```python
filtered_df["n video"] = filtered_df[columns_to_sum].sum(axis=1)
filtered_df["n quizz"] = filtered_df[["Quizz.1.bin", "Quizz.2.bin", "Quizz.3.bin", "Quizz.4.bin", "Quizz.5.bin"]].sum(axis=1)
```

```python
filtered_df.head()
```

```python
print(filtered_df["n video"].max())
print(filtered_df["n video"].min())
print(filtered_df["n quizz"].min())
print(filtered_df["n quizz"].max())
```

```python
grouped_df = filtered_df.groupby('Gender')['n video'].mean().reset_index()
grouped_df.head()
```

```python
filtered_df.groupby('Country_HDI')['n video'].mean().reset_index()
```

```python
mean_n_quizz_gender = filtered_df.groupby('Gender')['n quizz'].mean().reset_index()
mean_n_quizz_country_hdi = filtered_df.groupby('Country_HDI')['n quizz'].mean().reset_index()

print(mean_n_quizz_gender)
print(mean_n_quizz_country_hdi)
```

```python
grouped_hdi_quizz = filtered_df.groupby('Country_HDI')['n video'].agg(['mean', 'std']).reset_index()

plt.figure(figsize=(12, 6))
plt.bar(grouped_hdi_quizz['Country_HDI'], grouped_hdi_quizz['mean'], yerr=grouped_hdi_quizz['std'], color='skyblue', capsize=5)
plt.xlabel('Country HDI')
plt.ylabel('Average Number of Videos')
plt.title('Average Number of Videos by Country HDI')
plt.xticks(rotation=90)
plt.savefig("bar_plotHDI.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
grouped_gender_quizz = filtered_df.groupby('Gender')['n video'].agg(['mean', 'std']).reset_index()

plt.figure(figsize=(12, 6))
plt.bar(grouped_gender_quizz['Gender'], grouped_gender_quizz['mean'], yerr=grouped_gender_quizz['std'], color='skyblue', capsize=5)
plt.xlabel('Gender')
plt.ylabel('Average Number of Videos')
plt.title('Average Number of Videos by Gender')
plt.xticks(rotation=90)
plt.savefig("bar_plotGENRE.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
grouped_hdi_quizz = filtered_df.groupby('Country_HDI')['n quizz'].agg(['mean', 'std']).reset_index()

plt.figure(figsize=(12, 6))
plt.bar(grouped_hdi_quizz['Country_HDI'], grouped_hdi_quizz['mean'], yerr=grouped_hdi_quizz['std'], color='skyblue', capsize=5)
plt.xlabel('Country HDI')
plt.ylabel('Average Number of Quizzes')
plt.title('Average Number of Quizzes by Country HDI')
plt.xticks(rotation=90)
plt.show()
```

```python
grouped_gender_quizz = filtered_df.groupby('Gender')['n quizz'].agg(['mean', 'std']).reset_index()

plt.figure(figsize=(12, 6))
plt.bar(grouped_gender_quizz['Gender'], grouped_gender_quizz['mean'], yerr=grouped_gender_quizz['std'], color='skyblue', capsize=5)
plt.xlabel('Gender')
plt.ylabel('Average Number of Quizzes')
plt.title('Average Number of Quizzes by Gender')
plt.xticks(rotation=90)
plt.show()
```
